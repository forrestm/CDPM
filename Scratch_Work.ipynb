```python
test = shaping.ZV(0.2526851, 0.016807487)
test2 = shaping.EI(0.47542116, 0.20396849)
freq1, mag1 = shaping.sensplot(test.shaper,0,1,0.017)
freq2, mag2 = shaping.sensplot(test2.shaper,0,1,0.204)


# Set the plot size - 3x2 aspect ratio is best
fig = plt.figure(figsize=(6, 4))
ax = plt.gca()
plt.subplots_adjust(bottom=0.17, left=0.17, top=0.96, right=0.96)

# csfont = {'fontname':'CMU Serif'}
# Change the axis units to serif
plt.setp(ax.get_ymajorticklabels(), fontsize=18)
plt.setp(ax.get_xmajorticklabels(), fontsize=18)

# Remove top and right axes border
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Only show axes ticks on the bottom and left axes
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Turn on the plot grid and set appropriate linestyle and color
ax.grid(True,linestyle=':', color='0.75')
ax.set_axisbelow(True)
# plt.ylim(0,20)
plt.xlim(0,0.9)
ax.annotate(r'$V_{tol}$', xytext=(0.82, 3.5), xy=(0.82, 3.5), color='black', fontsize=20)
# Define the X and Y axis labels
plt.xlabel(r'Frequency (Hz)', fontsize=22, labelpad=5)
plt.ylabel(r'Percent Vibration', fontsize=22, labelpad=10)
plt.plot(freq1, mag1*mag2*100)
plt.plot([0,0.80],[5,5],linewidth=1, linestyle=':', label = r'5\% Allowable', color='black')
plt.savefig('Sensitivity_curve.pdf')
```
