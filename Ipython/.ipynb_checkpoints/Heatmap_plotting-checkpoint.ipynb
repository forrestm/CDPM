```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
XY_L1L2_natflmh_damplmh = np.genfromtxt('XY_L1L2_natflmh_damplmh2.csv', delimiter=',')
```

```python
import seaborn as sns
sns.set(context="paper", font="serif", font_scale=2)
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
sns.axes_style({'font.family': ['serif']})
```

```python
# plt.figure(0)
fig = plt.figure(figsize=(6, 4))
hea1 = np.column_stack((XY_L1L2_natflmh_damplmh[:,0],XY_L1L2_natflmh_damplmh[:,1]
                      ,XY_L1L2_natflmh_damplmh[:,4]/(2*np.pi)))
df1 = pd.DataFrame(hea1, columns=['X','Y','EL'])
Heats1 = df1.pivot("Y", "X", "EL")

ax1 = sns.heatmap(Heats1, xticklabels=False, yticklabels=False, annot=False, fmt="d",
            linewidths=0, cmap="inferno", cbar=False)
#             vmin=0,
#             vmax=6)
# plt.title('Low Mode Natural Frequency Across Workspace', y = 1.05)
cbar1 = ax1.figure.colorbar(ax1.collections[0])
# cbar1.ax.set_title(r'Hz', fontsize=22)

cbar1.set_label(r'Frequency (Hz)', fontsize=22)
plt.xlabel(r'Horizontal Position', fontsize=22)
plt.ylabel(r'Vertical Position', fontsize=22)
# plt.yticks(np.linalg())
# plt.xticks([0,5,10,15,19], rotation=360)
plt.tight_layout(pad=0.5)
plt.savefig("Low_Mode_Natural_Frequency.svg")
plt.show()
```

```python
# plt.figure(1)
fig = plt.figure(figsize=(6, 4))
hea2 = np.column_stack((XY_L1L2_natflmh_damplmh[:,0],XY_L1L2_natflmh_damplmh[:,1]
                      ,XY_L1L2_natflmh_damplmh[:,5]/(2*np.pi)))
df2 = pd.DataFrame(hea2, columns=['X','Y','EL'])
Heats2 = df2.pivot("Y", "X", "EL")
ax2 = sns.heatmap(Heats2, xticklabels=False, yticklabels=False, annot=False, fmt="d",
            linewidths=0, cmap="inferno", cbar=False)
#             vmin=0,
#             vmax=6)
# plt.title('Middle Mode Natural Frequency Across Workspace', y = 1.05)
cbar2 = ax2.figure.colorbar(ax2.collections[0])
# cbar2.ax.set_title('Hz', fontsize=22)
cbar2.set_label(r'Frequency (Hz)', fontsize=22)
plt.xlabel('Horizontal Position', fontsize=22)
plt.ylabel('Vertical Position', fontsize=22)
plt.tight_layout(pad=0.5)
plt.savefig("Middle_Mode_Natural_Frequency.svg")
plt.show()
```
